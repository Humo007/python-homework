```python
# #2. Person Class
# Write a Python program to create a Person class. Include attributes like name, country, and date of birth. Implement a method to determine the person's age.
from datetime import datetime
class Person:
    def __init__(self, name, country, date_of_birth):
        self.name =name
        self.country = country
        self.date_of_birth = datetime.strptime(date_of_birth, "%d.%m.%Y") 
    def age(self):
        today = datetime.today()
        age = today.year - self.date_of_birth.year

        if (today.month, today.day) < (self.date_of_birth.month, self.date_of_birth.day):
            age -= 1
        return age
    

if __name__ == "__main__":
    try:
        person1 = Person("Ali", "Uzbekistan", '13.09.2003' )
        print(f"{person1.name} is {person1.age()} years old.")
    except Exception as e:
        print(f"Error: {e}")
    
```

```python
# 3. Calculator Class
# Write a Python program to create a Calculator class. Include methods for basic arithmetic operations.

class Calculator:
    def __init__(self, number1, number2, operation):
        self.number1 = number1
        self.number2 = number2
        self.operation = operation
    def Calculate(self):
        if self.operation == '+':
            return self.number1 + self.number2
        elif self.operation == '-':
            return self.number1 - self.number2
        elif self.operation == "*":
            return self.number1 * self.number2
        elif self.operation == '/':
            if self.number2 == 0:
                return "Error: Division by zero"
            return self.number1 / self.number2
        else:
            return "Error: Unsupported operation"

first_operation = Calculator(12, 6, '*')
print(f"{first_operation.number1} {first_operation.operation} {first_operation.number2} = {first_operation.Calculate()}")


    
```

```python
# 4. Shape and Subclasses
# Write a Python program to create a class that represents a shape. 
# Include methods to calculate its area and perimeter. 
# Implement subclasses for different shapes like Circle, Triangle, and Square.
import math
class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement this method.")
    def perimeter(self):
        raise NotImplementedError("Subclass must implement this method.")
 

class Circile(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.radius


class Triangle(Shape):
    def __init__(self, side1, side2, side3):
        self.a = side1
        self.b = side2
        self.c = side3

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) *  (s - self.c))
class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def perimeter(self):
        return 4 * self.side
    
shapes = [Circile(5), Triangle(3, 4, 5), Square(4)]

for shape in shapes:
    print(f"{shape.__class__.__name__}: Area = {shape.area():.2f}, Perimeter = {shape.perimeter():.2f} ")
        
```

```python
# 6. Stack Data Structure
# Write a Python program to create a class representing a stack data structure. 
# Include methods for pushing and popping elements.

class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty!"
        

    def is_empty(self):
        return len(self.items) == 0
   
s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
print("Popping element: ", s.pop())   
```

```python
# 8. Shopping Cart Class
# Write a Python program to create a class representing a shopping cart.
# Include methods for adding and removing items, and calculating the total price.

class ShopingCard:
    def __init__(self):
        self.items = {}
    
    def add_items(self, name, price, quantity = 1):
        if name in self.items:
            current_price, current_qty = self.items[name]
            self.items[name] = (price, current_qty + quantity)
        else:
            self.items[name] = (price, quantity)
        print(f"{quantity} {name} added. ")


    def remove_item(self, name):
        if name in self.items:
            del self.items[name]
            print(f"The {name} was removed from the cart.")
        else:
            print(f"The {name} was not found in the cart.")

    def get_total(self):
        total = sum(price * qty for price, qty in self.items.values())
        return total
    
    def show_cart(self):
        if not self.items:
            print("ðŸ›’Cart is empty!")
        else:
            print("ðŸ“‹Products in the cart: ")
            for name, (price, qty) in self.items.items():
                print(f"{qty} {name} cost {price}/each. ") 


cart = ShopingCard()

while True:
    print("\n1. Add products.")
    print("2. Remove products.")
    print("3. Show total price.")
    print("4. Show products.")
    print("5. Exit.")
    
    choice = input("Choose an option (1-5): ")
    if choice == '1':
        name = input("Enter product name: ")
        try:
            price = float(input("Enter price: "))
            quantity = int(input("Enter quantity: "))
            cart.add_items(name, price, quantity)
        except ValueError:
            print("Please enter valid number values for price and quantity.")
    elif choice == '2':
        name = input("Enter product name to remove: ")
        cart.remove_item(name)
        
    elif choice == '3':
        print("ðŸ’° Total price: ", cart.get_total(),"$")

    elif choice == '4':
        cart.show_cart()
    elif choice == '5':
        print("Exiting the shopping card. Goodbye!")
        break
    else:
        print("Invalid choice. Please choose 1 to 5.")

            



```

```python
# 9. Stack with Display
# Write a Python program to create a class representing a stack data structure.
# Include methods for pushing, popping, and displaying elements.
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty!"
        

    def is_empty(self):
        return len(self.items) == 0
    
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
             return "Stack is empty!"
        
    def display(self):
        return self.items


s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)

print("Stack :", s.display())
print("The peek element: ", s.peek())
print("Popping element: ", s.pop())
print("After Stack : ", s.display())
    
```

```python
# 11. Bank Class
# Write a Python program to create a class representing a bank. 
# Include methods for managing customer accounts and transactions.

class Account:
    def __init__(self, account_number, balance=0.0):
        self.account_number = account_number
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"Deposited: ${amount}")
            return True
        return False

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            self.transactions.append(f"Withdrawn: ${amount}")
            return True
        return False

    def get_balance(self):
        return self.balance

    def get_transactions(self):
        return self.transactions


class Customer:
    def __init__(self, name, customer_id):
        self.name = name
        self.customer_id = customer_id
        self.accounts = {}

    def add_account(self, account_number, initial_balance=0.0):
        if account_number not in self.accounts:
            self.accounts[account_number] = Account(account_number, initial_balance)
            return True
        return False

    def get_account(self, account_number):
        return self.accounts.get(account_number)


class Bank:
    def __init__(self, name):
        self.name = name
        self.customers = {}

    def add_customer(self, customer_id, name):
        if customer_id not in self.customers:
            self.customers[customer_id] = Customer(name, customer_id)
            print(f"Customer '{name}' added.")
            return True
        print("Customer already exists.")
        return False

    def create_account(self, customer_id, account_number, initial_balance=0.0):
        customer = self.customers.get(customer_id)
        if customer:
            return customer.add_account(account_number, initial_balance)
        return False

    def deposit(self, customer_id, account_number, amount):
        customer = self.customers.get(customer_id)
        if customer:
            account = customer.get_account(account_number)
            if account:
                return account.deposit(amount)
        return False

    def withdraw(self, customer_id, account_number, amount):
        customer = self.customers.get(customer_id)
        if customer:
            account = customer.get_account(account_number)
            if account:
                return account.withdraw(amount)
        return False

    def check_balance(self, customer_id, account_number):
        customer = self.customers.get(customer_id)
        if customer:
            account = customer.get_account(account_number)
            if account:
                return account.get_balance()
        return None

    def print_transactions(self, customer_id, account_number):
        customer = self.customers.get(customer_id)
        if customer:
            account = customer.get_account(account_number)
            if account:
                for t in account.get_transactions():
                    print(t)


bank = Bank("KapitalBank")

while True:
    print("\n1. Add customer")
    print("2. Create account")
    print("3. Deposit")
    print("4. Withdraw")
    print("5. Check balance")
    print("6. Print transactions")
    print("7. Exit")

    choice = input("Choose an option (1-7): ")

    if choice == '1':
        customer_id = input("Enter customer ID: ")
        name = input("Enter customer name: ")
        bank.add_customer(customer_id, name)

    elif choice == '2':
        customer_id = input("Enter customer ID: ")
        account_number = input("Enter account number: ")
        try:
            initial_balance = float(input("Enter initial balance: "))
        except ValueError:
            print("Invalid amount.")
            continue
        if bank.create_account(customer_id, account_number, initial_balance):
            print("Account created.")
        else:
            print("Account already exists or customer not found.")

    elif choice == '3':
        customer_id = input("Enter customer ID: ")
        account_number = input("Enter account number: ")
        try:
            amount = float(input("Enter amount: "))
        except ValueError:
            print("Invalid amount.")
            continue
        if bank.deposit(customer_id, account_number, amount):
            print("Deposit successful.")
        else:
            print("Deposit failed.")

    elif choice == '4':
        customer_id = input("Enter customer ID: ")
        account_number = input("Enter account number: ")
        try:
            amount = float(input("Enter amount: "))
        except ValueError:
            print("Invalid amount.")
            continue
        if bank.withdraw(customer_id, account_number, amount):
            print("Withdrawal successful.")
        else:
            print("Withdrawal failed.")

    elif choice == '5':
        customer_id = input("Enter customer ID: ")
        account_number = input("Enter account number: ")
        balance = bank.check_balance(customer_id, account_number)
        if balance is not None:
            print(f"ðŸ’° Balance: ${balance}")
        else:
            print("Account not found.")

    elif choice == '6':
        customer_id = input("Enter customer ID: ")
        account_number = input("Enter account number: ")
        print("Transactions:")
        bank.print_transactions(customer_id, account_number)

    elif choice == '7':
        print("Exiting the Bank. Goodbye ðŸ‘‹")
        break

    else:
        print("Invalid choice. Please choose 1 to 7.")
```
