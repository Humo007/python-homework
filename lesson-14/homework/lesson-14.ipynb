```python
# 3. Task: JSON Modification
# Write a program that allows users to add new books, update existing book information,
# and delete books from the books.json JSON file.
import json
File_name = "books.json"
def save_books(books):
    with open(File_name, 'w', encoding="utf-8") as f:
        json.dump(books, f, indent=2, ensure_ascii=False )

def add_books(books_list:list):
    new_id = max([book['id'] for book in books_list ], default=0)+1
    title = input("Title: ")
    author = input("Author: ")
    year = int(input("Year: "))
    genre = input("Genre: ")
    books_list.append({"id": new_id, "title": title, "author": author, "year": year, "genre": genre})
    save_books(books_list)
    print("‚úÖ Book added successfully.")

def update_book(book_list: list):
    book_id = int(input("Enter the ID of the book to update: "))
    for book in book_list:
        if book['id'] == book_id:
            book['title'] = input(f"New title (current: {book['title']}): ") or book['title']

            book['author'] = input(f"New author (current: {book['author']}): ") or book['author']
            book['year'] = int(input(f"New year (current: {book['year']}): ")) or book['year']
            book['genre'] = input(f"New genre (current: {book['genre']}): ") or book['genre']
            save_books(book_list)
            print("‚úÖ Book updated.")
            return
    print("‚ùå Book not found")

def delete_book(book_list:list):
    book_id = int(input("Enter the ID of the book to delete: "))
    update_books = [book for book in book_list if book['id'] != book_id ]
    if len(book_list) == len(update_books):
        print("‚ùåBook not found!")
    else:
        save_books(update_books)
        print("‚úÖ Book deleted.")
def get_info(books_list):
    for i in books_list:
        print(i)


with open(File_name, 'r', encoding="utf-8") as f:
    books = json.load(f)
    while True:
        print("\nüìö Book maneger menu: ")
        print("1. Add book.")
        print("2. Update book.")
        print("3. Delete book.")
        print("4. Get info.")
        print("5. Exit.")
        choice = input("Enter an option: ")
        if choice == '1':
            add_books(books)
        elif choice == '2':
            update_book(books)
        elif choice == '3':
            delete_book(books)
        elif choice == '4':
            get_info(books)
        elif choice == '5':
            break
        else:
            print("‚ùåInvalid choice. Try again!")

```

```python
# 4. Task: Movie Recommendation System
# Use this url http://www.omdbapi.com/ to fetch information about movies.
# Create a program that asks users for a movie genre and recommends a random movie from that genre.
import requests
import random

api_key = "2f78eb03"

def search_movies_by_genre(genre):
    url =  f"http://www.omdbapi.com/?apikey={api_key}&s={genre}&type=movie"
    response = requests.get(url)
    if response.status_code == 200:
        data:dict = response.json()
        if data.get("Search"):
            return data["Search"]
        else:
            print("üòí No movies found for this genre.")
            return []
    else:
        print("‚ö†Ô∏è Failed to fetch date from OMDb API.")
        return []

def recommend_random_movie(movie):
    movie = random.choice(movie)
    print("\n üìΩÔ∏è Recommended Movie:")
    print(f"Title: {movie['Title']}")
    print(f"Year: {movie['Year']}")
    print(f"IMDB ID: {movie['imdbID']}")
    print(f"Type: {movie['Type']}")
    print(f"Poster: {movie['Poster']}")

def main():
    genre = input("Enter a movie genre (e.g. Action, Comedy, Horror): ")
    movies = search_movies_by_genre(genre)
    if movies:
        recommend_random_movie(movies)
main()
```
