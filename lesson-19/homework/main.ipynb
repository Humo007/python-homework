```python
# Homework Assignment 1: Analyzing Sales Data

# You are given a dataset containing sales data for an e-commerce website. The dataset (task\sales_data.csv) has the following columns:

# Date: Date of the sale.
# Product: Name of the product sold.
# Category: Category to which the product belongs.
# Quantity: Number of units sold.
# Price: Price per unit.
# Tasks:

```

```python
import pandas as pd
import numpy as np
df_sales = pd.read_csv("sales_data.csv")
df_sales.head()
```

```python
# 1.Group the data by the Category column and calculate the following aggregate statistics for each category:
    #Total quantity sold.
total_quantity = df_sales.groupby("Category")['Quantity'].agg('sum')
print('Total quantity sold: \n', total_quantity)
    #Average price per unit.
df_sales['Sales'] = df_sales['Quantity'] * df_sales['Price']
average_price = df_sales.groupby("Category").apply(lambda x: x['Sales'].sum()/x['Quantity'].sum())
print('Average price per unit: \n', average_price)
    #Maximum quantity sold in a single transaction.
max_quantity = df_sales.groupby("Category")['Quantity'].agg('max')
print('Maximum quantity sold in a single transaction: \n', max_quantity)
```

```python
# 2.Identify the top-selling product in each category based on the total quantity sold.
grouped = df_sales.groupby(['Category', 'Product'])['Quantity'].sum().reset_index()
top_selling_product = grouped.sort_values('Quantity', ascending=False).groupby('Category').first().reset_index()
top_selling_product
```

```python
# 3.Find the date on which the highest total sales (quantity * price) occurred.
df_sales['Sales'] = df_sales['Quantity'] * df_sales['Price']
dayly_sales = df_sales.groupby("Date")['Sales'].sum().reset_index()
# print(dayly_sales)
max_sales_date = dayly_sales.loc[dayly_sales["Sales"] == dayly_sales['Sales'].max()]
max_sales_date
```

```python
# Homework Assignment 2: Examining Customer Orders

# You have a dataset (task\customer_orders.csv) containing information about customer orders. The dataset has the following columns:

# OrderID: Unique identifier for each order.
# CustomerID: Unique identifier for each customer.
# Product: Name of the product ordered.
# Quantity: Number of units ordered.
# Price: Price per unit.
# Tasks:
import pandas as pd
df_customer_orders = pd.read_csv("customer_orders.csv")
df_customer_orders.head()
```

```python
# 1. Group the data by CustomerID and filter out customers who have made less than 20 orders.
customer_orders = df_customer_orders.groupby("CustomerID").size()
filter_customers = customer_orders[customer_orders >= 20]
df_filtered = df_customer_orders[df_customer_orders['CustomerID'].isin(filter_customers.index)]
df_filtered

#or
filtered_customers = df_customer_orders.groupby('CustomerID').filter(lambda x: len(x) >= 20)
print(filtered_customers)
```

```python
# 2. Identify customers who have ordered products with an average price per unit greater than $120.
df_customer_orders['Sales'] = df_customer_orders['Price'] * df_customer_orders['Quantity']
customer_average_price = df_customer_orders.groupby('CustomerID', group_keys=False).apply(lambda x: x['Sales'].sum() / x['Quantity'].sum())
high_avg_price_customer = customer_average_price[customer_average_price > 120].index
df_filtered_customer = df_customer_orders[df_customer_orders['CustomerID'].isin(high_avg_price_customer)]
print(df_filtered_customer)
```

```python
# 3. Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.
df_customer_orders['Sales'] = df_customer_orders['Price'] * df_customer_orders['Quantity']
product_status = df_customer_orders.groupby('Product').agg({
    'Quantity': 'sum',
    "Sales": 'sum'
}).reset_index()
filtered_products = product_status[product_status['Quantity'] >= 5]
filtered_products
```

```python
# Homework Assignment 3: Population Salary Analysis

# 1. "task\population.db" sqlite database has population table.
import sqlite3
import pandas as pd
conn = sqlite3.connect("population.db")
query = "SELECT * FROM population"
df_population = pd.read_sql_query(query, conn)
print(df_population)
```

```python
# 2. "task\population salary analysis.xlsx" file defines Salary Band categories.
    # Read the data from population table and calculate following measures:
df_population_salary = pd.read_excel("population_salary_analysis.xlsx")
df_population_salary.head()
    # Percentage of population for each salary category;
    # Average salary in each salary category;
    # Median salary in each salary category;
    # Number of population in each salary category;
def extract_range(band):
    band = str(band).strip()
    if "till" in band.lower():
        max_val = int(band.replace("till $", "").replace(",", "").strip())
        return (0, max_val)
    elif "and over" in band.lower():
        min_val = int(band.lower().replace("and over", "").replace("$", "").replace(",", "").strip())
        return (min_val, float('inf'))
    else:
        parts = band.replace("$", "").replace(",", "").split(" - ")
        return (int(parts[0]), int(parts[1]))



df_population_salary['Range'] = df_population_salary['Salary Band'].apply(extract_range)
df_population_salary
def assign_band(salary):
    for i, (low, high) in enumerate(df_population_salary['Range']):
        if low <= salary <= high:
            return df_population_salary.loc[i, 'Salary Band']
    return "Other"
 
df_population['Salary Band'] = df_population['salary'].apply(assign_band)
grouped = df_population.groupby('Salary Band')
result = grouped['salary'].agg([
    ("Average Salary", 'mean'),
    ("Median Salary", "median"),
    ("Number of population", 'count')
])
total = len(df_population)
result['Percentage'] = (result["Number of population"] / total * 100).round(2)
result = result.reset_index()
result['Range'] = result['Salary Band'].map(
    dict(zip(df_population_salary['Salary Band'], df_population_salary['Range']))

)
result = result.sort_values(by='Range').drop(columns='Range').reset_index()
result = result.drop(columns='index')
result
```

```python
# 3. Calculate the same measures in each State
# Note: Use SQL only to select data from database. All the other calculations should be done in python.
state_status = df_population.groupby('state')['salary'].agg(
    count = 'count',
    mean = 'mean',
    median = 'median',
    min = 'min',
    max = 'max',
    std = 'std'
    ).reset_index()
state_status

```
